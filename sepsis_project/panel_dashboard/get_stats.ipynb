```python
csv_file = "../FormattedData/37.csv"
df2 = pd.read_csv(csv_file)

del df2["Unnamed: 0"]

df2
```

```python
i=0
counter = 0
counter2 = 0
rows = len(df2)

while i<rows-1:
    value1 = df2.iloc[i][0]
    gender = df2.iloc[i][8]
   
    if value1 == 0.0 or value1 == 0:
        if gender == 1.0 or gender == 1:
            counter+=1
        else:
            counter2+=1

    i+=1
    
print(counter)
print(counter2)
#number of different males
```

```python
i=0
counter = 0
counter2 = 0
rows = len(df2)
nineteen = 0
thirtynine = 0
fiftynine = 0
seventynine=0
eighty = 0

while i<rows-1:
    value1 = df2.iloc[i][0]
    age = df2.iloc[i][7]
   
    if value1 == 0.0 or value1 == 0:
        if age < 20:
            nineteen+=1
        elif age < 40:
            thirtynine +=1
        elif age < 60:
            fiftynine +=1
        elif age < 80:
            seventynine +=1
        elif age >=80:
            eighty +=1

    i+=1
    
print(nineteen)
print(thirtynine)
print(fiftynine)
print(seventynine)
print(eighty)
#number of different males
```
