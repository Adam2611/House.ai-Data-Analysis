```python
def extract_vitals(df1, i):
    p_id = df1.iloc[i][0]
    date = df1.iloc[i][1]
    time = df1.iloc[i][2]
    pulse = df1.iloc[i][4]
    rr = df1.iloc[i][5]
    o2sat = df1.iloc[i][6]
    temp = df1.iloc[i][7]
    
    # print (df1.iloc[i][3])
    # print (type(df1.iloc[i][3]))
    if isinstance(df1.iloc[i][3], str):
        pre_bp = df1.iloc[i][3]
        if type(pre_bp) == float and math.isnan(pre_bp)==False:
            manual_bp = str(pre_bp)
            x = manual_bp.split("/")
            if len(x)==2:
                bp1 = float(x[0])
                bp2 = float(x[1])
            else:
                bp1 = None
                bp2 = None
        elif type(pre_bp) == str:
            manual_bp = str(pre_bp)
            if "E" in manual_bp:
                manual_bp = manual_bp.replace("E", "")
                manual_bp = manual_bp.strip()
            x = manual_bp.split("/")
            if len(x)==2:
                bp1 = float(x[0])
                bp2 = float(x[1])
            else:
                bp1 = None
                bp2 = None
        else:
            bp1 = None
            bp2 = None

    elif df1.iloc[i][8] and df1.iloc[i][9] and math.isnan(df1.iloc[i][8])==False and math.isnan(df1.iloc[i][9])==False:
        bp1 = df1.iloc[i][8]
        bp2 = df1.iloc[i][9]

    else:
        bp1 = None
        bp2 = None
    

    # print (df1.iloc[i][8])
    # print (df1.iloc[i][9])
    # print (math.isnan(df1.iloc[i][8]))
    # print (math.isnan(df1.iloc[i][9]))
    
    p_vitals = [p_id, date, time, bp1, bp2, pulse, temp, rr, o2sat]
    return p_vitals
```

```python
i = 0
pt_identifier = []
date = []
time = []
bp1 = []
bp2 = []
pulse = []
temp = []
rr = []
o2sat = []
while i < rows:
    p_vitals = extract_vitals(df1, i)
    # df2 = df2.append({'pt_identifier': p_vitals[0], 'Date': p_vitals[1],
    #                   'Time': p_vitals[2], 'BP1': p_vitals[3], 'BP2': p_vitals[4],
    #                   'Pulse': p_vitals[5], 'Temp': p_vitals[6], 'RR': p_vitals[7],
    #                   'O2Sat': p_vitals[8]}, ignore_index=True)
    #print(p_vitals)
    pt_identifier.append(p_vitals[0])
    date.append(p_vitals[1])
    time.append(p_vitals[2])
    bp1.append(p_vitals[3])
    bp2.append(p_vitals[4])
    pulse.append(p_vitals[5])
    temp.append(p_vitals[6])
    rr.append(p_vitals[7])
    o2sat.append(p_vitals[8])
    i += 1
  
    
```

```python
mydict = {"pt_identifier":pt_identifier, "Date": date, "Time": time, "BP1": bp1, "BP2": bp2, "Pulse": pulse, "Temp":temp, "RR": rr, "O2Sat":o2sat}
 #'Time': p_vitals[2], 'BP1': p_vitals[3], 'BP2': p_vitals[4],
    #                   'Pulse': p_vitals[5], 'Temp': p_vitals[6], 'RR': p_vitals[7],
    #                   'O2Sat': p_vitals[8]}, ignore_index=True)
```
