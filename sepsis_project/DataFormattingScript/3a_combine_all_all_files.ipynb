```python
len(big_pt)
import numpy as np
age = []
gender = []
visited = False
```

```python
for i in range(len(big_pt)):
    visited = False
    for j in range(len(small_pt1)):
        if big_pt[i] == small_pt1[j]:
            visited = True
            age.append(df_master1.iloc[j][5])
            gender.append(df_master1.iloc[j][6])
            # df_all["Age_Number"][i]= df_master1.iloc[j][5]
            # df_all["Gender"][i] = df_master1.iloc[j][6]
            break
    if visited == False:
        for k in range(len(small_pt2)):
            if big_pt[i] == small_pt2[k]:
                visited = True
                age.append(df_master2.iloc[k][5])
                gender.append(df_master2.iloc[k][6])
                break
        if visited == False:
            age.append(np.nan)
            gender.append("")
```

```python
df_all["Age_Number"] = age
df_all["Gender"] = gender
```
