```python
import pandas as pd
#import modin.pandas as pd
```

```python
rows = len(df1)
rows
```

```python
df1.iloc[2][1] #row, column
```

```python
del df1["pt_identifier"]
```

```python
df1["new_date"] = pd.to_numeric(df1["new_date"])
```

```python
i = 1
times = []
while i<rows:
    time = df1.iloc[i][0]
    prev_time = df1.iloc[i-1][0]
    if time != 0:
        times.append(time-prev_time)
    i+=1
```

```python
average = sum(times)/len(times)
```
