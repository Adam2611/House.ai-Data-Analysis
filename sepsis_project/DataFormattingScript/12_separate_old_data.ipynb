```python
import pandas as pd
```

```python
df1 = pd.read_csv(
     '../FormattedData/37.csv'
)
```

```python
rows = len(df1)
rows
```

```python
print (df1.iloc[0][1])
```

```python
i=1
date_list = []
date_list.append(df1.iloc[0][0])
days = 3
time_threshold = days*1440
while i<rows-2:
    value1 = df1.iloc[i][0]
    value0 = df1.iloc[i-1][0] 

    if (value1-value0) > time_threshold:
        subtracted = df1.iloc[i][0]
        date_list.append(0.0)
        new_idx = i+1

        while new_idx<rows and df1.iloc[new_idx][0] != 0.0 and (df1.iloc[new_idx][0]-df1.iloc[new_idx-1][0])<time_threshold:
            changed_v = df1.iloc[new_idx][0]
            date_list.append(changed_v - subtracted)
            new_idx += 1
            i += 1
    else:
        date_list.append(value1)
    

    i+=1
    
```

```python
date_list[0:50]
```

```python
del df1["new_date"]
df1["new_date"] = date_list
```

```python
df1 = df1[['new_date', 'BP1', 'BP2', 'Pulse', 'Temp', 'RR', 'O2Sat',
       'Age_Number', 'Male', 'Female']]
```
