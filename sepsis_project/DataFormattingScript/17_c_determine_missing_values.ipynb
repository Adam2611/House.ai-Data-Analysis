```python
new_date = []
bp1 = []
bp2 = []
hr = []
temp = []
rr = []
o2sat = []
age = []
lactate = []
wbc = []
```

```python
i = 1


while i<rows:
    if df1.iloc[i][0] == -1:
        new_date.append(0)
    else:
        new_date.append(1)

    if df1.iloc[i][1] == -1:
        bp1.append(0)
    else:
        bp1.append(1)

    if df1.iloc[i][2] == -1:
        bp2.append(0)
    else:
        bp2.append(1)

    if df1.iloc[i][3] == -1:
        hr.append(0)
    else:
        hr.append(1)

    if df1.iloc[i][4] == -1:
        temp.append(0)
    else:
        temp.append(1)

    if df1.iloc[i][5] == -1:
        rr.append(0)
    else:
        rr.append(1)

    if df1.iloc[i][6] == -1:
        o2sat.append(0)
    else:
        o2sat.append(1)

    if df1.iloc[i][7] == -1:
        age.append(0)
    else:
        age.append(1)

    if df1.iloc[i][10] == -1:
        lactate.append(0)
    else:
        lactate.append(1)

    if df1.iloc[i][11] == -1:
        wbc.append(0)
    else:
        wbc.append(1)

    i+=1
```

```python
def compute_missing(alist):
    size = len(alist)
    count = 0
    for i in alist:
        if i == 0:
            count+=1
    print(float(count)/float(size))
```

```python
compute_missing(new_date)
compute_missing(bp1)
compute_missing(bp2)
compute_missing(hr)
compute_missing(temp)
compute_missing(rr)
compute_missing(o2sat)
compute_missing(age)
compute_missing(lactate)
compute_missing(wbc)
#is the average number of steps
```
